### Muestreo sistematico:

```python
def systematic_sampling(econdata, step):
  indexes = np.arange(0, len(econdata), step=step)
  systematic_sample = econdata.iloc[indexes]
  return systematic_sample

systematic_sample = systematic_sampling(econdata, 3)
systematic_sample.head()
```

## 3) Muestreo estratificado

Consiste en crear unidades homogeneas exclusivas, y a partir de ahi crear una muestra aleatoria. 

```python
econdata['strat'] = econdata['delegacion'] + ", " + econdata['tipo']
(econdata['strat'].value_counts()/len(econdata)).sort_values(ascending=False)
```

Nuestros datos dicen que la proporciòn es la siguiente:
1) Hoteles en Cauhtemoc 0.5

2) Museos en Cuauhtemoc 0.2

3) Hoteles en Venustiano Carranza 0.1

4) Mercados en Cuauhtémoc 0.1

5) Mercados en Venustiano Carranza 0.1

```python
def strat_data(econdata, strat_cols, strat_values, strat_prop, random_state=None):
  df_strat = pd.DataFrame(columns = econdata.columns)

  pos = -1
  for i in range (len(strat_values)):
    pos +=1
    if pos == len(strat_values) -1:
      ratio_len = len(econdata) - len(df_strat)
    else:
      ratio_len = int(len(econdata) * strat_prop[i])

    df_filtrado = econdata[econdata[strat_cols] == strat_values[i]]
    df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

    df_strat = pd.concat([df_strat, df_temp])
  return df_strat
```

```python
strat_values = list(econdata['strat'].unique())
strat_prop = [0.5, 0.2, 0.1, 0.1, 0.1]
df_strat = strat_data(econdata,'strat', strat_values, strat_prop, random_state=42)
```
